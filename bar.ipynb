```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["sun","mon","tue","wed","fri"])
y=np.array([10,20,30,40,50])
plt.bar(x,y,color='hotpink',linewidth=0.1,width=0.2) 
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.xlabel('week name')
plt.ylabel('sale')
plt.barh(x, y, height = 0.1)
plt.show()
```
