```python
import matplotlib.pyplot as plt
x=['sun','mon','tue','wed','thu','fri'] 
y=[10,32,12,22,14,10]
plt.xlabel('week name')
plt.ylabel('sale in a day')
plt.title('bakery sale')
plt.plot(x,y,linewidth=2,linestyle='--',color='r',marker='s',markersize=10)
plt.grid(True)
plt.legend()
plt.show()
```
